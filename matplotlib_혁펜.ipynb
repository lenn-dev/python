## 🔎 예시 1: **위치, 방향, 회전 – 회전 행렬 (Rotation Matrix)**

### 🧠 사용된 삼각함수 개념:

* `cos(θ)`와 `sin(θ)`는 **회전 변환**의 핵심
* 평면상에서 어떤 점을 θ만큼 회전시키려면 이 두 값으로 구성된 **회전 행렬** 사용

### 💻 코드 예시 (2D 회전):

```python
import numpy as np
import matplotlib.pyplot as plt

# 45도 회전
theta = np.pi / 4
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

vec = np.array([1, 0])  # x축 방향 벡터
rotated_vec = rotation_matrix @ vec

# 시각화
plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color='blue', label='original')
plt.quiver(0, 0, rotated_vec[0], rotated_vec[1], angles='xy', scale_units='xy', scale=1, color='red', label='rotated')
plt.xlim(-1, 1.5); plt.ylim(-1, 1.5)
plt.gca().set_aspect('equal')
plt.legend()
plt.title("2D Rotation Using sin/cos")
plt.grid()
plt.show()
```

### 🔧 AI 활용 분야:

* **강화학습**: 로봇이나 드론의 **회전 방향** 제어
* **컴퓨터 비전**: 객체 인식에서 회전된 물체 탐지
* **3D 모델링/AR**: 객체의 자세(pose) 계산

## 🔎 예시 2: **주기적인 입력값 인코딩 (Positional Encoding in Transformers)**

### 🧠 개념:

* **sin/cos 파형으로 위치 정보를 인코딩**
* Transformer 구조는 순서를 따로 갖고 있지 않기 때문에, 입력 토큰의 "위치" 정보를 벡터에 **주기 함수로 추가**함

### 💻 코드 예시 (Positional Encoding 생성):

```python
import numpy as np

def positional_encoding(pos, d_model):
    PE = np.zeros((pos, d_model))
    for i in range(d_model):
        if i % 2 == 0:
            PE[:, i] = np.sin(np.arange(pos) / (10000 ** (i / d_model)))
        else:
            PE[:, i] = np.cos(np.arange(pos) / (10000 ** ((i - 1) / d_model)))
    return PE

pe = positional_encoding(50, 16)  # 50개의 위치, 차원 16
```

### 🔧 AI 활용 분야:

* **자연어처리(NLP)**: GPT, BERT, T5 등 모든 Transformer 모델
* **시계열 예측**: 순서 정보를 입력에 반영
* **멀티모달 학습**: 비디오 프레임 간 순서 부여

## 🔎 예시 3: **푸리에 변환 (Fourier Transform)**

### 🧠 개념:

* 푸리에 변환은 모든 신호를 **sin과 cos의 조합으로 표현**
* 음성, 이미지, 센서 데이터 등 다양한 입력에서 **주파수 성분 추출**

### 💻 코드 예시:

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500)
signal = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)

fft = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(signal), d=1/500)

plt.plot(freqs[:250], np.abs(fft[:250]))
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

### 🔧 AI 활용 분야:

* **음성 인식**: 음성 → 스펙트럼 변환 후 모델 입력
* **의료 AI**: 심전도(ECG), 뇌파(EEG) 분석
* **비전**: 이미지 압축, 텍스처 인식, 고속 처리

## 🔎 예시 4: **원형 특성 (Circular Data) 처리**

### 🧠 개념:

* 시간(시계 방향), 각도(0\~360도)는 선형이 아니라 **원형 데이터**야
* 이걸 **sin/cos로 변환하면 선형 모델도 학습 가능**

### 💻 예시 (시간을 sin/cos로 인코딩):

```python
import numpy as np
import matplotlib.pyplot as plt

hours = np.arange(24)
sin_hour = np.sin(2 * np.pi * hours / 24)
cos_hour = np.cos(2 * np.pi * hours / 24)

plt.plot(hours, sin_hour, label='sin(hour)')
plt.plot(hours, cos_hour, label='cos(hour)')
plt.legend()
plt.title("Hour as Circular Features")
plt.grid()
plt.show()
```

### 🔧 AI 활용 분야:

* **시간대별 사용자 행동 분석**
* **날씨 예측**, **전력 수요 예측**
* **교통량 분석**

## ✅ 정리: 삼각함수가 활용되는 AI 개념

| 삼각함수 개념      | 사용 분야      | 역할                  |
| ------------ | ---------- | ------------------- |
| `sin`, `cos` | 위치, 회전     | 회전 행렬 구성, 방향 계산     |
| 주기성          | 신호 처리, NLP | 주기적 입력 표현, 시계열 예측   |
| 주파수          | 음성, 센서     | 푸리에 변환 → 스펙트럼 분석    |
| 원형 데이터 변환    | 시계, 계절성    | sin/cos로 선형 공간으로 매핑 |
